`````python

# Collect temperatures for 5 days
temperatures = []
for day in range(1, 6):
    temp = float(input(f"Enter temperature for Day {day} (°C): "))
    temperatures.append(temp)

# Write to file
with open("temps.txt", "w") as f:
    for day, temp in enumerate(temperatures, 1):
        f.write(f"Day {day}: {temp}°C\n")

# Read file and calculate average
with open("temps.txt", "r") as f:
    lines = f.readlines()

print("\n--- Temperature Log ---")
total = 0
for line in lines:
    print(line.strip())
    # Extract numeric temperature value
    temp_val = float(line.split(": ")[1].replace("°C", ""))
    total += temp_val

average = total / len(lines)
print(f"\nAverage Temperature: {average:.1f}°C")
`````

### Activity 15: Managing Study Groups

You are responsible for managing a study group for a course. Each student can be part of multiple study groups, and you need to track which groups a student is enrolled in. Students can join or leave study groups. What data structure would you choose to manage this efficiently, and why?

**Example data:**

```python
{"John": ["Math", "Physics"], "Emily": ["Physics", "Biology"], "Alex": ["Math"]}

Steps:

1. Choose a data structure that allows quick access to the groups each student belongs to.
2. Write a program that lets students join and leave study groups.
3. Implement a function to display all groups a student is part of.
---



`````python

# Best data structure: Dictionary of Lists
# Key = student name (string), Value = list of groups
# Allows O(1) lookup per student, easy add/remove groups

study_groups = {
    "John":  ["Math", "Physics"],
    "Emily": ["Physics", "Biology"],
    "Alex":  ["Math"]
}

def join_group(student, group):
    """Add a student to a group."""
    if student not in study_groups:
        study_groups[student] = []
    if group not in study_groups[student]:
        study_groups[student].append(group)
        print(f"{student} joined {group}.")
    else:
        print(f"{student} is already in {group}.")

def leave_group(student, group):
    """Remove a student from a group."""
    if student in study_groups and group in study_groups[student]:
        study_groups[student].remove(group)
        print(f"{student} left {group}.")
    else:
        print(f"{student} is not in {group}.")

def display_groups(student):
    """Display all groups a student belongs to."""
    groups = study_groups.get(student, [])
    if groups:
        print(f"{student}'s groups: {', '.join(groups)}")
    else:
        print(f"{student} is not in any group.")

# Demo
join_group("Alex", "Biology")
leave_group("John", "Physics")
display_groups("Alex")
display_groups("John")
`````

### Activity 20: Managing Event Tickets

You are developing a system to manage event tickets for concerts or sports events. Each event has a limited number of tickets available, and users can purchase tickets if they are still available. You need to track which users have purchased tickets for specific events and how many are still available. What data structure would you use for this task, and why?

**Example data:**
- Available tickets: `{"Concert": 100, "Football Game": 50, "Theater Show": 75}`
- Purchasers: `{"Concert": ["John", "Emily"], "Football Game": ["Alex"], "Theater Show": ["Sam", "John"]}`

**Steps:**
1. Choose a data structure to track ticket availability and purchasers efficiently.
2. Write a program that handles ticket purchases and updates the list of purchasers.
3. Implement functions to check availability and view who has purchased tickets.
---



`````python


# - Outer key: event name
# - Value: dict with "available" (int) and "purchasers" (list)

ticket_system = {
    "Concert":       {"available": 100, "purchasers": ["John", "Emily"]},
    "Football Game": {"available": 50,  "purchasers": ["Alex"]},
    "Theater Show":  {"available": 75,  "purchasers": ["Sam", "John"]}
}

def check_availability(event):
    """Check how many tickets are left for an event."""
    if event in ticket_system:
        print(f"{event}: {ticket_system[event]['available']} ticket(s) available.")
    else:
        print(f"Event '{event}' not found.")

def purchase_ticket(user, event):
    """Purchase a ticket if available."""
    if event not in ticket_system:
        print(f"Event '{event}' not found.")
        return
    ev = ticket_system[event]
    if ev["available"] > 0:
        ev["available"] -= 1
        ev["purchasers"].append(user)
        print(f"{user} purchased a ticket for {event}. {ev['available']} left.")
    else:
        print(f"Sorry, no tickets left for {event}.")

def view_purchasers(event):
    """View everyone who bought a ticket."""
    if event in ticket_system:
        print(f"Purchasers for {event}: {ticket_system[event]['purchasers']}")

# Demo
check_availability("Concert")
purchase_ticket("Sara", "Concert")
view_purchasers("Concert")
purchase_ticket("Tom", "Football Game")
check_availability("Football Game")

`````
