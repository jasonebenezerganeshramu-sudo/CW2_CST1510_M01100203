### Typical 1. Total Time to Watch a Series
Compute total viewing time (no breaks) for a series:
- 5 seasons, 10 episodes per season
- Season 1: 45 minutes 35 seconds per episode
- Seasons 2–5: 35 minutes 15 seconds per episode
Display the total as hours, minutes, and seconds.
**Example Output:**
```
Total time: 39h 30m 0s
```


````python
# Your code here
no_season = 5
no_episods = 10
s1_min, s1_sec = 45, 35
s2_min, s2_sec = 35, 15

s1_time_sec = ((45 * 60) + 35) * 10
s2_time_sec = ((35 * 60) + 15) * (no_episods * 4)

total_sec = s1_time_sec + s2_time_sec

hours = total_sec // (3600)   
remainder = total_sec % (3600)
minutes = remainder // 60
sec = remainder % 60
print(f'{hours}:{minutes}:{sec}')
````

### Typical 2. Savings Projection (3 Years)
Compute final savings after 3 years given:
- Save £10 every week for 3 years
- Year 1 bonus: +£20 each week
- Every 2 weeks: −£0.50 (sister)
Assume 52 weeks per year; use arithmetic and augmented assignment; no conditionals.
**Example Output:**
```
Final savings after 3 years: £4689.50
```


````python
'''
year 1 = 10/ week 52
year 2 = 20/ week 52
year 3 = 20 /week 52
 total weeks   = 156
'''

weeks_year = 52
weeks_3_year = weeks_year * 3
savings_in_3_years = weeks_3_year * 10
savings_with_bonus = (weeks_year * 2 )  * 10
sister_money = (weeks_3_year /2) * 0.5
total_savings = savings_in_3_years + savings_with_bonus 
total = total_savings - sister_money
print(f'{total_savings} - sister money {sister_money} = {total}')
````
