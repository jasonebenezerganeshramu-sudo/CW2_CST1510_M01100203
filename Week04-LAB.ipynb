### Task 6: Classify Server Log Entries

Write a function `classify_logs(logs)` that takes a list of log strings and:
1. Counts how many `ERROR`, `INFO`, and `WARNING` entries there are
2. Extracts and prints the **timestamp** from each entry
3. Returns the three counts

**Methods to use:** `startswith()`, `split()`, `strip()`


```python
logs = [
    "[14:32:05] ERROR: Disk full",
    "[14:32:10] INFO: Server started",
    "[14:33:01] WARNING: High memory usage",
    "[14:33:15] INFO: Backup complete",
    "[14:34:00] ERROR: Connection refused",
    "[14:35:22] WARNING: CPU at 90%"
]

def classify_logs(logs):
    error_count = 0
    info_count = 0
    warning_count = 0

    # TODO: Loop through each log entry
    # Extract the timestamp (between the square brackets)
    # Check if the text after the timestamp starts with ERROR, INFO, or WARNING
    # Increment the appropriate counter
    # Print each log with its timestamp

    # Your code here
    for entry in logs:
        # extract timestamp between the square brackets
        timestamp = entry[1:entry.index("]")]
        # get the text after "] "
        after = entry[entry.index("]") + 2:].strip()

        if after.startswith("ERROR"):
            error_count += 1
            level = "ERROR"
        elif after.startswith("WARNING"):
            warning_count += 1
            level = "WARNING"
        else:
            info_count += 1
            level = "INFO"

        print(f"[{timestamp}] {level}")

    return error_count, info_count, warning_count

errors, infos, warnings = classify_logs(logs)
print(f"\nErrors: {errors}, Info: {infos}, Warnings: {warnings}")
```

### Task 7: Simple Sentiment Analyser

Write a function `analyse_sentiment(reviews)` that:
1. Cleans each review (`strip()`, `lower()`)
2. Classifies it as **positive**, **negative**, or **neutral** using keyword lists
3. Returns a dictionary with counts and a list of which reviews are negative

**Methods/Functions to use:** `strip()`, `lower()`, `any()`, `in`


```python
reviews = [
    "  AMAZING product, Love it!!!  ",
    "terrible experience   ",
    "  It was okay  ",
    "WORST purchase ever",
    "  fantastic quality  ",
    "poor customer service",
    "  Not bad, not great  "
]

positive_words = ['amazing', 'love', 'fantastic', 'great', 'excellent']
negative_words = ['terrible', 'worst', 'poor', 'awful', 'horrible']

def analyse_sentiment(reviews):
    # Your code here
    counts = {'positive': 0, 'negative': 0, 'neutral': 0}
    negative_reviews = []

    for review in reviews:
        clean = review.strip().lower()
        if any(word in clean for word in positive_words):
            counts['positive'] += 1
        elif any(word in clean for word in negative_words):
            counts['negative'] += 1
            negative_reviews.append(review.strip())
        else:
            counts['neutral'] += 1

    counts['negative_reviews'] = negative_reviews
    return counts

result = analyse_sentiment(reviews)
print(result)
```

### Task 8: Data Validation Pipeline

Write a function `validate_record(record)` that takes a CSV string and checks:
1. It has exactly 4 fields (name, age, email, phone)
2. Name contains only letters and spaces
3. Age is a valid number between 18 and 100
4. Email contains `@` and `.`
5. Phone starts with `+` or `0`

Return a list of **error messages** (empty list = valid record).

Use `all()` to check if ALL fields are non-empty first.

**Methods to use:** `split()`, `isdigit()`, `isalpha()`, `startswith()`, `all()`


```python
records = [
    "Alice Smith,25,alice@email.com,+447123456789",
    "Bob,abc,bob@email,07123456789",
    "Charlie123,30,charlie@mail.com,447123456789",
    ",22,missing@mail.com,+440000000000"
]

def validate_record(record):
    errors = []
    # Your code here
    fields = record.split(",")

    if len(fields) != 4:
        errors.append("Must have exactly 4 fields")
        return errors

    name, age, email, phone = fields

    # Check all fields are non-empty
    if not all(f.strip() for f in fields):
        errors.append("One or more fields are empty")

    # Name: only letters and spaces
    if not all(c.isalpha() or c == ' ' for c in name.strip()):
        errors.append("Name must contain only letters and spaces")

    # Age: valid number between 18 and 100
    if not age.strip().isdigit() or not (18 <= int(age.strip()) <= 100):
        errors.append("Age must be a number between 18 and 100")

    # Email: must contain @ and .
    if '@' not in email or '.' not in email:
        errors.append("Email must contain '@' and '.'")

    # Phone: must start with + or 0
    if not phone.strip().startswith(('+', '0')):
        errors.append("Phone must start with '+' or '0'")


for record in records:
    errors = validate_record(record)
    if errors:
        print(f" {record}")
        for e in errors:
            print(f"   - {e}")
    else:
        print(f"{record}")
```

### Task 10: Password Strength Checker

Write a function `check_password(password)` that scores a password (0–5):

| Rule | Points |
|------|--------|
| At least 8 characters | +1 |
| Contains uppercase letter | +1 |
| Contains lowercase letter | +1 |
| Contains a digit | +1 |
| Contains a special character (`@#$%!`) | +1 |

Print the score and a label: **Weak** (0-2), **Medium** (3-4), **Strong** (5).

If the password contains a space, print  and reject it.

**Methods to use:** `len()`, `isupper()`, `islower()`, `isdigit()`, `any()`


```python
test_passwords = [
    "hello",
    "Hello123",
    "My Pass!",
    "Str0ng@Pass",
    "12345678"
]

def check_password(password):
    # Your code here
    if ' ' in password:
        print(f"'{password}' → REJECTED (contains a space)")
        return

    score = 0
    if len(password) >= 8:                      score += 1
    if any(c.isupper() for c in password):      score += 1
    if any(c.islower() for c in password):      score += 1
    if any(c.isdigit() for c in password):      score += 1
    if any(c in '@#$%!' for c in password):     score += 1

    label = "Weak" if score <= 2 else ("Medium" if score <= 4 else "Strong")
    print(f"'{password}' → Score: {score}/5 ({label})")

for pw in test_passwords:
    check_password(pw)
```
