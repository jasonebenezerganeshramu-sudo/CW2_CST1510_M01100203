### Example Task 2: Student Marks Analysis (Threshold — Pure Python EDA)

Stores a list of exam marks and computes the average, highest, and lowest marks, then counts how many students passed (>=50) versus failed, using list functions and a generator expression.

```python
marks   = [45, 78, 62, 90, 34, 55, 88, 40, 71, 66]

average = sum(marks) / len(marks)
highest = max(marks)
lowest  = min(marks)
passed  = sum(1 for m in marks if m >= 50)
failed  = len(marks) - passed

print(f'Marks:   {marks}')
print(f'Average: {average:.2f}')
print(f'Highest: {highest}')
print(f'Lowest:  {lowest}')
print(f'Passed:  {passed}   Failed: {failed}')
```

### Example Task 1: Celsius/Fahrenheit Converter (Example Activity)

Defines two conversion functions and lets the user choose a direction before converting an entered value between Celsius and Fahrenheit.

```python
def cel_to_fah(celsius):
    fahrenheit = (celsius * 1.8) + 32
    return fahrenheit

def fah_to_cel(fahrenheit):
    celsius = (fahrenheit - 32) / 1.8
    return celsius

choice = input('Enter:\n1. to convert from Celsius to Fahrenheit\n2. to convert from Fahrenheit to Celsius\n > ')
value = float(input('Enter the value you would like to convert: '))

result = cel_to_fah(value) if choice == '1' else fah_to_cel(value)
print(f'The result from the conversion is: {result}')
```

### Example Task 2: Matrix Row Sum (Threshold)

```python
def sum_row(matrix, rowIndex):
    return sum(matrix[rowIndex])

matrix = [[1, 3, 2], [10, 20, 2], [10, 30, 5]]

print(sum_row(matrix, 0))
```

### Example Task 2: Password Strength Checker with Scoring (Typical, Task 10)

Scores a password out of 5 based on length, uppercase, lowercase, digit, and special-character rules, then labels it Weak, Medium, or Strong (and rejects passwords containing spaces).

```python
def check_password(password):
    if ' ' in password:
        print(f"'{password}' -> REJECTED (contains a space)")
        return

    score = 0
    if len(password) >= 8:                  score += 1
    if any(c.isupper() for c in password):  score += 1
    if any(c.islower() for c in password):  score += 1
    if any(c.isdigit() for c in password):  score += 1
    if any(c in '@#$%!' for c in password): score += 1

    label = "Weak" if score <= 2 else ("Medium" if score <= 4 else "Strong")
    print(f"'{password}' -> Score: {score}/5 ({label})")
```

### Example Task 1: Personal Profile File

Writes name, age, and favourite colour to my_profile.txt, one value per line, then reads the file back and prints each value with a descriptive label.

```python
with open("my_profile.txt", "w") as f:
    f.write("Alex\n")
    f.write("20\n")
    f.write("Blue\n")

labels = ["Name", "Age", "Favorite Color"]
with open("my_profile.txt", "r") as f:
    lines = f.readlines()

for label, line in zip(labels, lines):
    print(f"{label}: {line.strip()}")
```

### Example Task 2: Simple Grade Book Dictionary 

Manages a dictionary of student scores: adds a new student, updates an existing score, prints everyone scoring above 80, and calculates the class average.

```python
grades = {"Alice": 85, "Bob": 92, "Charlie": 78}

grades["Diana"] = 88
grades["Bob"] = 95

print("Students with scores above 80:")
for student, score in grades.items():
    if score > 80:
        print(f"  {student}: {score}")

average = sum(grades.values()) / len(grades)
print(f"\nClass average score: {average:.1f}")
```
